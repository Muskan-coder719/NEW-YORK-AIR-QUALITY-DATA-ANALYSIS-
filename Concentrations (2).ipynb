```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets for each year
data_2019 = pd.read_csv('CO 2019.csv')
data_2020 = pd.read_csv('CO 2020.csv')
data_2021 = pd.read_csv('CO 2021.csv')

# Combine datasets into a single DataFrame
data = pd.concat([data_2019, data_2020, data_2021])

# Convert 'Date' to datetime and extract 'Year'
data['Date'] = pd.to_datetime(data['Date'])
data['Year'] = data['Date'].dt.year

# Aggregate data by Local Site Name and Year
site_year_avg = data.groupby(['Local Site Name', 'Year'])['Daily Max 8-hour CO Concentration'].mean().reset_index()

# Plot the data
plt.figure(figsize=(14, 8))
for year in site_year_avg['Year'].unique():
    subset = site_year_avg[site_year_avg['Year'] == year]
    plt.plot(subset['Local Site Name'], subset['Daily Max 8-hour CO Concentration'], marker='o', label=str(year))

plt.title('Average Daily Max 8-hour CO Concentration by Local Site')
plt.xlabel('Local Site Name')
plt.ylabel('CO Concentration (ppm)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.legend(title='Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
ozone_data = pd.read_csv('ozone_2019_cleaned_data.csv')
co_data = pd.read_csv('CO 2019.csv')
no2_data = pd.read_csv('NO2 2019.csv')

# Ensure 'Date' column is in datetime format
ozone_data['Date'] = pd.to_datetime(ozone_data['Date'])
co_data['Date'] = pd.to_datetime(co_data['Date'])
no2_data['Date'] = pd.to_datetime(no2_data['Date'])

# Group by month and calculate mean concentrations for each pollutant
monthly_ozone = ozone_data.groupby(ozone_data['Date'].dt.month)['Daily Max 8-hour Ozone Concentration'].mean()
monthly_co = co_data.groupby(co_data['Date'].dt.month)['Daily Max 8-hour CO Concentration'].mean()
monthly_no2 = no2_data.groupby(no2_data['Date'].dt.month)['Daily Max 1-hour NO2 Concentration'].mean()

# Combine the data into a single DataFrame
monthly_data = pd.DataFrame({
    'Ozone': monthly_ozone,
    'CO': monthly_co,
    'NO2': monthly_no2
})

# Create a figure with two y-axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar plot for Ozone and CO
monthly_data[['Ozone', 'CO']].plot(kind='bar', ax=ax1, position=0, color=['blue', 'red'], width=0.8)
ax1.set_ylabel('Concentration (ppm)', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=0)
ax1.set_title('Average Monthly Concentrations of Ozone, CO, and NO2 for year 2019')

# Create a second y-axis for NO2
ax2 = ax1.twinx()
monthly_data['NO2'].plot(kind='line', ax=ax2, color='green', marker='o', linewidth=2, label='NO2')
ax2.set_ylabel('NO2 Concentration (ppb)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Combine the legends from both axes
ax1.legend(['Ozone', 'CO'], loc='upper left')
ax2.legend(['NO2'], loc='upper right')

plt.tight_layout()
plt.show()
```
